## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique = len(purchase_data['Item ID'].unique())
purchase = len(purchase_data['Item ID'])
avg_price = purchase_data['Price'].mean()
revenue = purchase_data['Price'].sum()

summary = pd.DataFrame({'Unique Items': [unique], 
                        'Average Price': [avg_price], 
                        'Number of Purchases': [purchase], 
                        'Total Revenue': [revenue]})
summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
non_duplicates = purchase_data.drop_duplicates(subset='SN')

Total_count = non_duplicates['Gender'].count()
male_count = non_duplicates['Gender'].value_counts()['Male']
female_count = non_duplicates['Gender'].value_counts()['Female']
other_nd_count = non_duplicates['Gender'].value_counts()['Other / Non-Disclosed']

male_percent = (male_count/Total_count)*100
female_percent = (female_count/Total_count)*100
other_nd_percent = (other_nd_count/Total_count)*100

summary_gen = pd.DataFrame({'Male Count': [male_count],
                          'Female Count': [female_count],
                          'Other / Non-Dislcosed Count': [other_nd_count],
                           'Male Percent': [male_percent],
                           'Female Percent': [female_percent],
                           'Other / Non-Disclosed Percent': [other_nd_percent]})
summary_gen



```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_data = purchase_data.groupby(['Gender'])

gen_count = gender_data['Price'].count()
gen_avg_price = gender_data['Price'].mean()
gen_total = gender_data['Price'].sum()
gen_avg_person = gen_total / gen_count

gen_sum = pd.DataFrame({'Purchase Count': gen_count,
                       'Average Purchase Price': gen_avg_price,
                       'Total Purchase Value': gen_total,
                       "Avg Total Purchase per Person": gen_avg_person})

gen_sum
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 1000]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

bins_data = purchase_data.copy()
bins_data['Group'] = pd.cut(bins_data['Age'], bins, labels=labels)
bins_group = bins_data.groupby(['Group'])

bins_count = bins_group['Purchase ID'].count()
bins_percent = (bins_count / purchase)*100

dem_table  = pd.DataFrame({'Total Count': bins_count,
                          'Percentage of Players': bins_percent})

dem_table
```
